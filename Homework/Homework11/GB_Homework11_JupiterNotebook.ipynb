# Дана функция: f(x) = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
```

```python
x = Symbol('x')
function = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
function
```

### 1. Определение корней:

```python
solveset(function)
```

### 2. Нахождение интервала, на которых функция возрастает:

Не разобрался как делать

```python
pro = diff(function) 
pro
```

```python
grafic = plot(pro, function, (x, -10, 10))
```

### 4. График функции:

```python
grafic = plot(function, (x, -10, 10))
```

### 5. Вершина(ы):

```python
y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
print(f'x = {x}, y ={y}')
print(f'x = {x.evalf()}, y = {y.evalf()}')
```

### 6. Определить промежутки, на котором f > 0

```python
pprint(solve(function > 0, x), use_unicode=True)
```

### 7. Определить промежутки, на котором f < 0

```python
pprint(solve(function < 0, x), use_unicode=True)
```
